## **Case Study – Shopping Cart System**
### **Instructions:**

1. Create a list named cart with initial items: Milk, Bread, Eggs, Apple

2. Perform list operations:

* Add multiple items using extend()

 * Insert item at index 2 using insert()

*  Remove item Apple using remove()

* Remove last item using pop()

3. Use list methods:

* index() to find position of Eggs

* count() to count occurrences of Milk

4. Apply indexing:

* Display first item

* Display last item

5. Apply slicing:

* Display first 3 items from the list

6. Find total number of items using len()

7. Display all outputs using print()

```python
# Case Study – Shopping Cart System

# Creating shopping cart list
cart = ["Milk", "Bread", "Eggs", "Apple"]

print("Initial Cart:", cart)

# extend() - Add multiple items
cart.extend(["Banana", "Butter"])
print("\nAfter extend:", cart)

# insert() - Insert item at index 2
cart.insert(2, "Cheese")
print("\nAfter insert:", cart)

# remove() - Remove specific item
cart.remove("Apple")
print("\nAfter remove:", cart)

# pop() - Remove last item
cart.pop()
print("\nAfter pop:", cart)

# index() - Find position of Egg
print("\nIndex of Eggs:", cart.index("Eggs"))

# count() - Count occurrences of Milk
print("Count of Milk:", cart.count("Milk"))

# indexing
print("\nFirst item:", cart[0])
print("Last item:", cart[-1])

# slicing
print("\nFirst 3 items:", cart[:3])

# length
print("\nTotal items in cart:", len(cart))
```

## **Case Study – Student Marks Analysis**
### **Instructions:**

Create a list named marks by getting 5 subject marks from the user using map() and list()

Perform calculations:

* Find total marks using sum()

* Find average marks using

* Calculate over all percentage assuming each subject is out of 100

```python
# Create a list named marks by getting 5 subject marks from the user
marks = list(map(int, input("Enter 5 subject marks separated by space: ").split()))

# Find total marks
total = sum(marks)

# Find average marks
average = total / len(marks)

# Calculate overall percentage
percentage = (total / (len(marks) * 100)) * 100

# Display results
print("Total Marks:", total)
print("Average Marks:", average)
print("Overall Percentage:", percentage, "%")
```
