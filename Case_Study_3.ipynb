# **Case Study – Student Details using Tuple (Packing & Unpacking)**
## **Instructions:**

1. Create a tuple named student and store the following details: name, age, course

2. Unpack the tuple into separate variables: name, age, course

3. Display each value separately using print().

```python
# Packing and Unpacking Tuples

# Packing (multiple values into a single tuple)
student = ("John", 22, "Data Analytics")

print("Packed Tuple:", student)

# Unpacking (extracting values from tuple into variables)
name, age, course = student

print("Name:", name)
print("Age:", age)
print("Course:", course)
```

# **Case Study – Online Order Summary using Tuple**
## **Instructions:**

1. Create a tuple named order with the following details: order_id, product_name, quantity, price, status

2. Unpack the tuple into separate variables: order_id, product, quantity, price, status

3. Display all order details using print() and f-string

```python
# Packing order details into tuple
order = ("ORD101", "Laptop", 1, 55000, "Delivered")

# Unpacking tuple
order_id, product, quantity, price, status = order

# Display details
print("----- Order Summary -----")

print("Order ID  :", order_id)
print("Product   :", product)
print("Quantity  :", quantity)
print("Price     :", price)
print("Status    :", status)
```
