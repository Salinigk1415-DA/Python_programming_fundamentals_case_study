## **Case Study – Student Performance Management System**
### **Instructions**
1. Create two lists: students marks
2. Use range() to: Access and print student names and marks using index positions
3. Use enumerate() to: Display student index number and student name
4. Use zip() to:
* Combine student names and marks together
* Display all outputs using for loops
* Use proper headings with print() statements
* Print the final result in a clean format

```python
# Student Marks System

students = ["John", "Sam", "David"]
marks = [90, 85, 95]

print("----- Using range() -----")

for i in range(len(students)):
    print(students[i], "-", marks[i])

print("----- Using enumerate() -----")

for index, name in enumerate(students):
    print(index, "-", name)

print("----- Using zip() -----")

for name, score in zip(students, marks):
    print(name, "-", score)
```
