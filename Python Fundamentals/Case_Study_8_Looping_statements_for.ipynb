## **1.Case Study – Text Analyzer System**
### **Instructions:**
1. Take a string input from the user using input().
2. Convert the text into lowercase using lower().
3.Use a for loop to iterate through each character in the string.
4. Count:
* Total vowels
* Total consonants
* Total digits
* Total special characters
5. Use:
* isalpha() for alphabets
* isdigit() for numbers
6. Display all counts using print()

```python
# Case Study – Text Analyzer System

text = input("Enter the string :").lower() #Good Morning

v = c = d = sc = 0

for letter in text:

  if letter in "aeiou":
    v +=1
  elif letter.isalpha():
    c +=1
  elif letter.isdigit():
    d +=1
  else:
    sc += 1

print(v,c,d,sc)
```

## **2. Case Study – Restricted Word Detection System**
### **Instructions:**
1. Create a list of words containing normal and restricted words.
2. Use a for loop to read words one by one.
3. Check whether the current word is a restricted word using if condition.
4. If the restricted word is found:
* Display "Restricted word detected"
* Stop the loop using break
5. Otherwise display all safe words using print()




```python
# Case Study – Restricted Word Detection System

# List of words
words = ["Hello", "Python", "Admin", "Blocked", "Welcome"]

# Restricted word
restricted_word = "Blocked"

# for loop
for word in words:

    # Checking restricted word
    if word == restricted_word:
        print("Restricted word detected")
        break

    # Safe words
    print(word)# Case Study – Restricted Word Detection System
```

## **3. Case Study – Employee Attendance Count System using continue**
### **Instructions:**
1. Create a list named attendance with values: Present, Absent, Present, Leave, Present
2. Create count variables:
* present
* absent
* leave
3. Use a for loop to iterate through the attendance list.
4. Use continue statement to:
* Skip remaining conditions after counting Present employees
5. Count:
* Total Present employees
* Total Absent employees
* Total Leave employees
6. Display all counts using print()

```python
# Case Study – Employee Attendance Count System using continue

# Attendance list
attendance = ["Present", "Absent", "Present", "Leave", "Present"]

# Count variables
present = 0
absent = 0
leave = 0

# for loop
for status in attendance:

    # Count present employees
    if status == "Present":
        present += 1
        continue

    # Count absent employees
    if status == "Absent":
        absent += 1

    # Count leave employees
    elif status == "Leave":
        leave += 1

# Display counts
print("Total Present:", present)
print("Total Absent:", absent)
print("Total Leave:", leave)
```
