## **Case Study – Arithmetic Calculator using User Defined Function**
### **Instructions**

Create a function called arth_cal()

Pass two parameters to the function

Inside the function:

Find addition

Find subtraction

Find multiplication

Return all three values using multiple return statements

Take two numbers from the user using input()

Store returned values in separate variables

Display:

Sum

Difference

Multiplication

```python
# Arithmetic Calculation using User Input

def arth_cal(a, b):

    s = a + b
    d = a - b
    m = a * b

    return s, d, m


# User Input
num1 = int(input("Enter First Number : "))
num2 = int(input("Enter Second Number : "))

# Function Call
sum, difference, multiplication = arth_cal(num1, num2)

# Output
print("\nSum of 2 numbers :", sum)
print("Difference of 2 numbers :", difference)
print("Multiplication of 2 numbers :", multiplication)
```

## **Case Study – Student Marks Analysis using Function and Loop**
### **Instructions**

Create a list called marks

Create a function called marks_analysis() and pass the marks list to the function

Inside the function:

Use a for loop to display all marks

Find total marks

Find average marks

Return total and average marks, call the function, and display: Student Marks, Total Marks, Average Marks

```python
# Marks List
marks = [80, 75, 90, 85, 70]

# Function Creation
def marks_analysis(data):

    print("Student Marks :")

    # for loop
    for i in data:
        print(i)

    # Operations
    total = sum(data)
    average = total / len(data)

    # Return Values
    return total, average


# Function Call
total_marks, average_marks = marks_analysis(marks)

# Output
print("\nTotal Marks   :", total_marks)
print("Average Marks :", average_marks)
```
