## **Case Study: Bank Loan Eligibility System**
### **Instructions:**

1.Create variables: customer_name, age, income, credit_score, existing_loan

2. Use conditional statements (if, elif, else) to:
* Approve loan if age ≥ 21, income ≥ 30000, credit_score ≥ 700, and no existing loan
* Conditionally approve loan if credit_score ≥ 600 and income ≥ 40000
* Otherwise reject the loan application

3. Display customer details and loan status using print()

```python
# Step 1: Customer Data
customer_name = "Ravi"
age = 28
income = 55000
credit_score = 720
existing_loan = False

# Step 2: Loan Eligibility Check
if age >= 21 and income >= 30000:
    if credit_score >= 700 and not existing_loan:
        status = "Approved "
    elif credit_score >= 600 and income >= 40000:
        status = "Conditionally Approved "
    else:
        status = "Rejected "
else:
    status = "Not Eligible "

# Step 3: Display Result
print("Customer:", customer_name)
print("Age:", age)
print("Income:", income)
print("Credit Score:", credit_score)
print("Loan Status:", status)
```

## **Case Study Scholarship Eligibility Checker**
### **Instructions:**

1. Create variables: student_name, marks, attendance, family_income.
2. Use conditional statements (if, elif, else) to:
* Award Full Scholarship if marks ≥ 90 and attendance ≥ 90.
* Award Half Scholarship if marks ≥ 80 or attendance ≥ 85.
* Award Need-Based Scholarship if marks ≥ 70 and family_income < 30000.
* Otherwise, display Not Eligible.
3. Display all data (no loops).

```python
student_name = "Ananya"
marks = 88
attendance = 92
family_income = 28000

if marks >= 90 and attendance >= 90:
    scholarship = "Full Scholarship "
elif marks >= 80 or attendance >= 85:
    scholarship = "Half Scholarship "
elif marks >= 70 and family_income < 30000:
    scholarship = "Need-Based Scholarship "
else:
    scholarship = "Not Eligible "

print("Student:", student_name)
print("Marks:", marks)
print("Attendance:", attendance)
print("Family Income:", family_income)
print("Scholarship Status:", scholarship)
```
