## **Case Study – String Analysis using List Comprehension**
### **Instructions**

Create a string variable called text

Store any sentence or word inside the variable

Use list comprehension to:

Convert all characters into uppercase

Extract only vowels from the string

Extract only consonants from the string

Display:

Original string

Uppercase characters

Vowels list

Consonants list

String without spaces

Print the output in a clean format using print() statements

```python
# String Analysis using List Comprehension

text = "data analytics"

# Convert to uppercase
upper_text = [x.upper() for x in text]

# Extract vowels
vowels = [x for x in text if x in "aeiou"]

# Extract consonants
consonants = [x for x in text if x.isalpha() and x not in "aeiou"]

# Remove spaces
remove_spaces = [x for x in text if x != " "]

print("Uppercase Letters :", upper_text)
print("Vowels            :", vowels)
print("Consonants        :", consonants)
print("Without Spaces    :", remove_spaces)
```

## **Case Study – Employee Salary Increment using List Comprehension**
### **Instructions**

Create a list called salary

Store employee salaries inside the list

Use list comprehension to: Increase each salary by 10%

Create another list to: Display only salaries greater than 50000

Print:

Original salary list

Updated salary list

Filtered salary list

```python
# Employee Salary Increment System

salary = [25000, 40000, 55000, 60000, 75000]

# Increment salary by 10%
updated_salary = [i * 1.10 for i in salary]

# Filter salaries greater than 50000
high_salary = [i for i in updated_salary if i > 50000]

print("Original Salary :", salary)
print("Updated Salary  :", updated_salary)
print("High Salary     :", high_salary)
```
