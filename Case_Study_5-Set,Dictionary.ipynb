## **Case Study – Student Course Registration System using Set**
## **Instructions:**

1. Create a set named courses with course names: Python, SQL, Power BI, Python, Excel

2. Perform set operations: • Add course using add()
* Add multiple courses using update()
* Remove course using remove() and discard()

3. Display all courses and total number of courses using len()

```python
# Case Study – Student Course Registration System using Set

# Creating set
courses = {"Python", "SQL", "Power BI", "Python", "Excel"}

# Display original set
print("Original Courses:", courses)

# add() - Add one course
courses.add("Tableau")
print("\nAfter add():", courses)

# update() - Add multiple courses
courses.update(["AI", "Machine Learning"])
print("\nAfter update():", courses)

# remove() - Remove course
courses.remove("SQL")
print("\nAfter remove():", courses)

# discard() - Remove safely
courses.discard("Java")
print("\nAfter discard():", courses)

# Display total courses
print("\nTotal Courses:", len(courses))
```

## **Case Study – Student Data Management System**
### **Instructions:**

1. Create a dictionary named student with the following keys: name, age, course

2. Store multiple students' data using lists as values.

3. Add one new student details to the dictionary:

* Append new name to name list
* Append new age to age list
* Append new course to course list

4. Display all student details using dictionary keys.

5. Access and display:
* First student's details using indexing
* Newly added student's details using indexing

```python
# Case Study – Student Data Management System

student = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [21, 22, 23],
    "course": ["Data Science", "AI", "Machine Learning"]
}

# Add one new student
student["name"].append("David")
student["age"].append(24)
student["course"].append("Python")

# Display all data
print("All Names:", student["name"])
print("All Ages:", student["age"])
print("All Courses:", student["course"])

# First student details
print("\n----- First Student -----")
print("Name:", student["name"][0])
print("Age:", student["age"][0])
print("Course:", student["course"][0])

# Second student details
print("\n----- Second Student -----")
print("Name:", student["name"][1])
print("Age:", student["age"][1])
print("Course:", student["course"][1])

# Newly added student details
print("\n----- New Student -----")
print("Name:", student["name"][3])
print("Age:", student["age"][3])
print("Course:", student["course"][3])
```
